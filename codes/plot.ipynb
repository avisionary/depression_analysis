```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
# import altair as alt

depression=pd.read_csv("../data/depression_overall.csv")
```

```python
depression['AgeBand'] = pd.cut(depression['age'], 4)
depression[['AgeBand','target']].groupby(['AgeBand'], as_index=False).mean().sort_values(by='AgeBand', ascending=True)
# Changing names of columns to remove ',' and brackets
#depression['AgeBand'].replace({'(12.924, 32.0]': 'AgeBand_1','(32.0, 51.0]':'AgeBand_2',
                           # '(51.0, 70.0]':'AgeBand_3','(70.0, 89.0]':'AgeBand_4'}, inplace=True,columns = None)

# Combining target categories
depression['target'] = np.where(
     depression['target'].between(-1, 3, inclusive=False), 0, 1)
```

```python
depression_plot = depression.drop(['TIPI1', 'TIPI2', 'TIPI3', 'TIPI4', 'TIPI5', 'TIPI6', 'TIPI7', 'TIPI8',
       'TIPI9', 'TIPI10','Condition','age','VCL'],axis=1,inplace=False)
```

```python
hist2 = alt.Chart(fig_data).mark_bar().encode(
    x='count',
    y='target',
    color = alt.Color('label_var',scale=alt.Scale(scheme="yellowgreenblue"),legend=None),
    tooltip=[alt.Tooltip("target", title="target"),
            alt.Tooltip("label_var", title="label_var"),
            alt.Tooltip("demo", title="demo")]  
).add_selection(
     selection
 ).transform_filter(
   selection
).properties(width=400, height=100)
hist2
```

```python
total_participation_viz=alt.vconcat(hist1, hist2).properties(
    title={
      "text": ["Which Demographic by Label has the most count?"], 
      "subtitle": ["Click to see Count of Demographic by Depression Target"],
      "subtitleFontSize":13,
      'subtitleFontStyle':'italic',
      "color": "black",
      "subtitleColor": "black"
    }
).configure(background='#E7E6E1')
total_participation_viz
```
