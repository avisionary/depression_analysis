```python
import pandas as pd
import matplotlib.pyplot as plt


depression_us=pd.read_csv("../data/depression_us.csv")
```

```python


multi_cols=['education', 'urban', 'gender', 'religion', 'orientation',
       'race', 'married']


#Duplicating columns for multi value columns
depression_us = pd.get_dummies(data = depression_us, columns = multi_cols)

```

```python
depression_us.drop('Condition',axis=1,inplace=True)
```

```python
depression_us
```

```python
depression_us=depression_us.drop(depression_us.columns[0], axis=1)
```

```python
depression_targets=depression_us['target']
depression_targets=depression_targets.astype('category')
depression_us.drop('target',axis=1,inplace=True)
```

```python
depression_us['age']-=depression_us['age'].mean()
depression_us['age']/=depression_us['age'].std()
```

```python
depression_us['familysize']-=depression_us['familysize'].mean()
depression_us['familysize']/=depression_us['familysize'].std()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Ridge, Lasso, LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor, BaggingRegressor, RandomForestRegressor, StackingRegressor
from xgboost import XGBRegressor
# from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(depression_us,depression_targets, test_size=0.25)
```

```python
lr = LogisticRegression(multi_class='multinomial')
pred_lr = lr.fit(X_train[['TIPI3','TIPI4','TIPI6','TIPI8','TIPI9']], y_train).predict(X_test[['TIPI3','TIPI4','TIPI6','TIPI8','TIPI9']])

# Checking Model Metrics
# from sklearn.metrics import r2_score, mean_squared_error
# print("R2 Score: ", r2_score(y_test, pred_lr))
# print("RMSE: ", np.sqrt(mean_squared_error(y_test, pred_lr)))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(splitter='best')

pred_dt = dt.fit(X_train[['TIPI3','TIPI4','TIPI6','TIPI8','TIPI9']], y_train).predict(X_test[['TIPI3','TIPI4','TIPI6','TIPI8','TIPI9']])
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred_dt))
```
